# Introducción

También llamada teoría de grafos o teoría de redes, la teoría de graficas es una rama de matemáticas discretas que estudia las relaciones entre parejas de elementos y cómo estas influyen en el comportamiento del conjunto.

Son de gran utilidad pues permiten modelar una gran cantidad de fenómenos o sistemas en la vida real:
- Flujo de Información
- Intercambios Financieros
- Relaciones Sociales
- Esparcimiento de Epidemias
- Interacciones en Ecosistemas
- Transporte y Cadenas de Suministro

Definiremos los elementos básicos de una gráfica:


1. **Vércites o Nodos:** todos los puntos en la gráfica. Forman el conjunto $V=\{v_1,v_2,...,v_n\}$.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-17-17-36-24.png">

2. **Aristas:** las conexiones entre dos vértices de la gráfica. En inglés se les llama *edges*. Son un conjunto $E \subseteq \{(u,v)\in V^2 \}$

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-14-12-29.png">

3. **Gráfica:** el *par ordenado* $G=(V,E)$. Aquí hay 2 gráficas diferentes con el mismo $V$.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-15-34-56.png">

Las aristas $(u,v)$ pueden ser de distintos tipos y eso da origen a distintos tipos de gráficas.


1. **Gráfica No-Dirigida:** las aristas son pares *no-ordenados*, por lo que no tienen asociada una dirección.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-14-21-361.png">

2. **Gráfica Dirigida:** las aristas son pares *ordenados*, por lo tienen una dirección. También se les llama flechas o líneas dirigidas.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-15-20-39.png">

3. **Gráfica Simple:** las aristas en $E$ son únicas.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-14-23-26.png">

4. **Gráfica Múltiple:** 2 vertices se pueden unir por más de una arista. Formalmente, en este caso se define una función $\Phi: E → \{(x, y) \in V^2\}$ del conjunto de vértices $E$ al conjunto de pares de vértices.  

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-14-32-35.png">

5. **Pseudográfica:** Generalmente asumimos que una arista $(u,v)$ es tal que $u \neq v$, i.e. los vértices no se conectan a sí mismos. A las gráficas en donde sí se permite esto se les llama **pseudográficas**.

<img src="images/pseudograph.png">

Para esto utilizamos `graphviz`. Podemos crear la gráfica en `networkx` y convertirla a una en `graphviz`. Sin embargo, la forma de hacer la configuración es distinta.

```python
import networkx as nx
from networkx.drawing.nx_pydot import write_dot
graph= nx.MultiDiGraph()                                                                                    

nodelist=['A','B','C','D']
edgelist=[('A','A'),('A','B'),('B','A'),('B','C'),('C','D'),('A','D')]
node_positions={'A':(100,100),'B':(-100,100),'C':(-100,-100),'D':(100,-100)}

node_colors=['Blue','Red','Blue','Red']
edge_colors=['Blue','Red','Blue','Blue','Red','Blue']
#edge_labels={('A','B'):'AB1',('A','B'):'AB2',('B','C'):'BC',('C','D'):'CD',('A','D'):'AD'}

graph.add_nodes_from(nodelist)
graph.add_edges_from(edgelist)

# assign node properties                                                                                        
print("Configurando Vértices")
for i,n in enumerate(graph.nodes):                                                                                           
    print(n,'"%d,%d"'%node_positions[n])
    graph.node[n]['pos'] = '"%d,%d"'%node_positions[n]
    graph.node[n]['color'] = node_colors[i]
    graph.node[n]['shape'] = 'circle'
    graph.node[n]['style'] = 'filled'
    graph.node[n]['fontcolor'] = 'white'
    graph.node[n]['width'] = '0.75'
    #graph.node[n]['fixedsize'] = 'True'

    
print("Configurando Aristas")
ordered_colors=[color for edge,color in sorted(zip(edgelist,edge_colors))]
    
for i,e in enumerate(graph.edges):
    print(i,e) #las aristas se ordenan alfabéticamente, por lo que tenemos que reordenar los colores
    graph.edges[e]['color']= ordered_colors[i]
    graph.edges[e]['label']= '"%s%s"'%(e[0],e[1]) 

#convertir la gráfica a objeto dot (usado en graphviz)    
write_dot(graph,'graph.dot')
#nx.nx_pydot.write_dot(graph,'graph.dot')

# ! permite ejecutar código de la terminal desde el notebook
!neato -T -n2 png graph.dot > graph.png #convierte el achivo dot a png

import matplotlib.image as mpimg
plt.figure(figsize=(10,10))
img=mpimg.imread('graph.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

# Los 7 Puentes de Köningsberg

Consideremos el siguiente problema:

La ciudad de Köningsberg, Prusia (hoy Kaliningrado, Rusia) es cruzada por un río, el cual tiene 7 puentes que conectan las diferentes partes de la ciudad como se muestra en la imagen. ¿Existe una trayectoria que podamos tomar para caminar por la ciudad y cruzar los 7 puentes sólo una vez?

<img src="https://upload.wikimedia.org/wikipedia/commons/5/5d/Konigsberg_bridges.png">

**Preguna:** ¿Cómo podríamos plantear el problema utilizando gráficas?

**Respuesta:** Una forma de hacerlo es pensar que las masas de tierra son los vértices y los puentes son las aristas que las conectan.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-13-58-24.png">

Dada la gráfica, podemos hacer las siguientes observaciones:

- Debemos escoger una masa de tierra (vértice) para iniciar y una para terminar la caminata (pueden o no ser la misma).
- Para cada vértice que no sea de inicio o final, debemos poder entrar y salir cada vez que pasemos,i.e. el número de aristas adyacentes debe ser par.
- El número de aristas adyacentes a cada vértice es impar.

Por lo tanto **no es posible** hallar una solución al problema.

Hay varios conceptos que podemos extaer del problema:

- **Grado de un Vértice:** El número de aristas adyacentes al vértice $v$. Denotado $deg(v)$.

<img src="images/degree_text.png">

- **Caminata:** Una sucesión de vértices $(v_1,v_2,v_3,...,v_l)$ tales que $v_i$ y $v_{i+1}$ estén conectador por una arista, i.e. $(v_i,v_{i+1})\in E$, para $i=1,...,l-1$. Puede repetir vértices o aristas. Es **cerrada** si $v_1=v_l$.

- **Trayectoria:** Una caminata en la que no se repiten las aristas. Puede repetir vértices.

- **Camino:** Una caminata en la que no se repiten ni vértices ni aristas.

<img src="https://cdncontribute.geeksforgeeks.org/wp-content/uploads/Untitled-drawing-2-2.png" width=400>



**Pregunta:** Da un ejemplo de una caminata que repita aristas pero no vértices.

- **Circuito:** Es una trayectoria cerrada, i.e. $v_1=v_l$. Puede repetir vértices además del inicio/final.
- **Cíclo:** Es un camino cerrado. No repite vértices ni aristas, excepto el vértice de inicio/final una sola vez.

<img src="https://cdncontribute.geeksforgeeks.org/wp-content/uploads/graph-1-6.png" width=400>

- **Caminata Euleriana:** Una caminata que cruza cada arista de la gráfica exactamente una vez. Es una trayectoria. Si es cerrada, es un **circuito euleriano**.
- **Caminata Hamiltoniana:** Una caminata que cruza cada vértice de la gráfica exactamente una vez. Es un camino. Si es cerrado, es un **cíclo Hamiltoniano**.

<img src="https://media.geeksforgeeks.org/wp-content/uploads/Euler-circuit.jpg">

- **Árboles:** Son gráficas que no contienen ningún cíclo.

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-05-16-36-16.png">

Una forma útil de representar las adyacencias en una gráfica es mediante una **matriz de adyacencias**. La entrada $g_{ij}$ en una gráfica simple no-dirigida es 1 si los vértices $v_i$ y $v_j$ son adyacentes (comparten una arista). 

<img src="http://mathworld.wolfram.com/images/eps-gif/AdjacencyMatrix_1002.gif">

**Pregunta:** ¿Cómo se generaliza la matriz de adyacencia para los otros tipos de gráficas?

En Python, la matriz de adyacencias de una gráfica G se puede obtener fácilmente con las funciones `adjacency_matrix` y `to_numpy_matrix` de `networkx`. [[link]](https://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.linalg.graphmatrix.adjacency_matrix.html)

```python
import networkx as nx
from networkx.drawing.nx_pydot import write_dot
graph= nx.MultiDiGraph()                                                                                    

nodelist=['A','B','C','D']
edgelist=[('A','A'),('A','B'),('B','A'),('B','C'),('C','D'),('A','D')]
node_positions={'A':(100,100),'B':(-100,100),'C':(-100,-100),'D':(100,-100)}

node_colors=['Blue','Red','Blue','Red']
edge_colors=['Blue','Red','Blue','Blue','Red','Blue']

graph.add_nodes_from(nodelist)
graph.add_edges_from(edgelist)

print(graph.nodes,graph.edges)

import matplotlib.image as mpimg
plt.figure(figsize=(10,10))
img=mpimg.imread('graph.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
graph.nodes
```

```python
nx.adjacency_matrix(graph).todense() #matriz de scipy
```

```python
adj_mat=nx.to_numpy_matrix(graph) #matriz de numpy
adj_mat
```

Esta matriz tiene varias propiedades útiles:

1. Sumar sobre las columnas nos permite obtener el número de aristas que entran a los vértices.

```python
adj_mat.sum(axis=0)
```

2. Sumar sobre los renglones nos permite obtener el número de aristas que salen de los vértices.

```python
adj_mat.sum(axis=1)
```

3. Multiplicar la matriz por sí misma n-veces nos permite ver el número de caminatas posibles de longitud n entre un vértice y otro. 

```python
adj_mat*adj_mat*adj_mat
```

# Hallar el Camino Óptimo

El parque nacional Sleeping Giant en Conneticut se ilustra en el siguiente mapa:

<img src="https://i1.wp.com/www.hikingwithdogs.xyz/wp-content/uploads/2015/12/sleeping-giant.jpg">

Cada año se organiza una carrera entre dos extremos del parque. Sin embargo, no existe un único camino para llegar a la meta. Por lo tanto, surge la pregunta: ¿cómo podemos hallar el camino óptimo?

**Pregunta:** ¿Cuáles serían los vértices y las aristas en este problema?

No es suficiente representar la gráfica. Además de eso necesitamos poder incorporar la distancia de cada porción del camino óptimo.

Una modificación útil que se realiza a las gráficas es asociar cantidades a sus vértices o aristas. En particular una cantidad asociada a un vértice se denomina el **peso del vértice**. Una gráfica con pesos asociados a sus vértices se denomina **gráfica pesada**. 

<img src="https://s3-ap-south-1.amazonaws.com/av-blog-media/wp-content/uploads/2018/09/Screenshot-from-2018-09-12-15-54-15.png">

Los pesos generalmente representan cantidades como distancia o tiempo entre nodos, costo de translado o flujo de una cierta cantidad (gráficas dirigidas).

Con esto, podemos tener nociones de distancia más generales en gráficas.

- **Camino más corto entre 2 vértices:** El camino entre dos vértices $v,u$ tales que la suma de los pesos de las aristas en el camino sea mínima. Si no hay pesos en la gráfica, se asume que todos tienen valor 1.

- **Distancia entre 2 vértices:** La suma de los pesos de las aristas del camino más corto. Se denota $d(u,v)$. En gráficas dirigidas puede que $d(u,v)\neq d(v,u)$.

- **Excentricidad de un vértice:** La distancia máxima entre un vértice $v$ y cualquier otro en la gráfica.

### Pregunta:
a) ¿Cuál es la distancia entre A y E?

b) ¿Cuál es la excentricidad de A?

c) ¿Cuál es el vértice con menor excentricidad?

Esto nos permitirá resolver el problema del camino óptimo. Empezamos por crear la gráfica a partir de tablas de datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

nodelist = pd.read_csv('https://gist.githubusercontent.com/brooksandrew/f989e10af17fb4c85b11409fea47895b/raw/a3a8da0fa5b094f1ca9d82e1642b384889ae16e8/nodelist_sleeping_giant.csv')
print(nodelist.shape)
nodelist.head(5)
```

```python
# Creamos una gráfica vacía y le añadimos los datos de las tablas
g = nx.Graph()

print("Añadir Aristas y Propiedades")
for i, elrow in edgelist.iterrows():
    print(i,elrow[0],elrow[1],elrow[2:].to_dict())
    g.add_edge(elrow[0],elrow[1]) #añade vértices automáticamente
    g.edges[elrow[0],elrow[1]].update(elrow[2:].to_dict())
    
print("\n Añadir Propiedades de los Vértices")
for i, nlrow in nodelist.iterrows():
    print(nlrow[1:].to_dict())
    g.node[nlrow['id']].update(nlrow[1:].to_dict())
```

Supongamos que los extremos del parque son los nodos "b_end_west" y "b_end_east". ¿Cómo podemos hallar el camino óptimo entre estos nodos?

### Algoritmo de Dijkstra

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e4/DijkstraDemo.gif/220px-DijkstraDemo.gif">

Este algoritmo permite hallar el camino más corto entre un nodo de inicio y cualquier otro nodo en la gráfica. Consiste en:

1. Marcar todos los vértices como no-visitados.
2. Inicializar las distancias tentativas: 0 para el nodo de inicio e infinito para todos los demas.
3. Fijar el nodo actual en el nodo inicial.
4. Considerar todos los vecinos del nodo actual y recalcular las distancias tentativas: comparar las distancias tentativas con los valores actuales y asignar el más pequeño de los dos. 
5. Marcar el nodo como visitado.
6. Fijar el nodo no-visitado con menor distancia tentativa como nodo actual.
7. Repetir los pasos 4 a 6 hasta llegar al nodo destino.

<img src="http://pi.math.cornell.edu/~mec/Winter2009/Thompson/search/dijkstra.png">

**Obs:** El algoritmo falla cuando hay un ciclo de aristas con pesos negativos.

En `networkx` hay 2 funcioes que sirven para implementar este algoritmo

```python
#longitud del camino óptimo entre 2 nodos
nx.dijkstra_path_length(g, 'b_end_west','b_end_east', weight='distance')
```

```python
#dictionario con nodos del camino óptimo dados el inicio y final
dict(nx.all_pairs_dijkstra_path(g, weight='distance'))['b_end_west']['b_end_east']
```

```python
nx.get_edge_attributes(g, 'distance')
```

**Pregunta:** ¿Cómo podrían los organizadores de la carrera elegir los puntos de inicio y fin, de modo que la distancia fuese máxima?

**Respuesta:** Deben fijarse los vértices de máxima excentricidad en la gráfica.

Hay varios conceptos que podemos extraer de esta pregunta:

- **Radio de la gráfica:** El mínimo de las excentricidades de los vértices en la gráfica.
- **Diametro de la gráfica:** El máximo de las excentricidades de los vértices de la gráfia.


- **Vértice central:** Vértice cuya excentricidad es igual al radio (es mínima).
- **Vértice periférico:** Vértice cuya excentricidad es igual al diametro (es máxima).

```python
distances=dict(nx.all_pairs_dijkstra_path_length(g,weight='distance'))
distances
```

```python
eccentricities=nx.distance_measures.eccentricity(g,sp=distances)
eccentricities
```

```python
nx.diameter(g,e=eccentricities)
```

Con esto, vemos que un vértice periférico es `b_end_east` y, al fijarnos en las distancias entre este vértice y todos los demás, vemos que el punto más lejano es `b_end_west` como habíamos supuesto.

```python
distances['b_end_east']
```

Sin embargo, este tipo de medida sólo considera los primeros vecinos a un nodo dado, i.e. una "cercanía de primer órden". Para estudiar la cercanía de distintos componentes de la gráfica se utilizan distintas medidas de centralidad. Veremos las más comunes: 

```python
G = nx.karate_club_graph()
pos = nx.spring_layout(G)
```

```python
#import matplotlib.colors as mcolors

def draw_graph(G, pos, measures, measure_name):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()), nodelist=measures.keys())
    #nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
    
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

**Degree Centrality**:La medida de centralidad más sencilla. El grado de cada vértice, aquí normalizado por el número total de vértices en la gráfica.

$$C_i=\frac{1}{n}\text{deg}(v_i)$$

```python
draw_graph(G, pos, nx.degree_centrality(G), 'Degree Centrality')
```

**Closeness Centrality:** El promedio de las distancias de un vértice a todos los demás.

$$C_i=\frac{1}{n}\sum_j d(v_i,v_j)$$

```python
draw_graph(G, pos, nx.closeness_centrality(G), 'Closeness Centrality')
```

**Betweeness Centrality:** El número de caminos más cortos que pasan por un cierto nodo, normalizado por el número de caminos más cortos entre cada par de nodos.

$$C_i=\sum_{j\neq i \neq k} \frac{\sigma_{jk}(i)}{\sigma_{jk}}$$

```python
draw_graph(G, pos, nx.betweenness_centrality(G), 'Betweenness Centrality')
```

**Eigenvector Centrality:** Esta es una medida un poco más complicada que generaliza la medida de centralidad de grado. Comenzamos definiendo la medida de centralidad tentativa de todos los nodos como $c_i=1$. Ahora iteramos sobre el nodo i utilizando: $c_i=\sum_j A_{ij}c_j$, o en notación de matrices:

$$\mathbf{c}(t)= \mathbf{A}^t \mathbf{c}(0)$$

```python
draw_graph(G, pos, nx.eigenvector_centrality(G), 'Eigenvector Centrality')
```
